```python
import requests
import pandas as pd
from datetime import datetime
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
access_token='xxx' 
headers = {'Authorization':"Token "+access_token}
```

```python
# loop through all pages to obtain all issues information
# label story point = 1
page_num = 1
issues_l=[]
for page_num in range(1,30):
    try:
        # to find all the repos' names from each page
        url = f"https://api.github.com/search/issues?q=is:issue+archived:false+is:public+label:%22Points:%201%22&page={page_num}"
        issue_l=requests.get(url,headers=headers).json()
        issue_l2=issue_l['items']
        issues_l.append(issue_l2)
    except:
        issues_l.append(None)
```

```python
repository=[]
for page in issues_l:
    for issue_l2 in page:
        try:
            repository.append(issue_l2['repository_url'])
        except:
            pass

len(repository)
```

```python
repository_data=pd.DataFrame()
repository_data['repo_url']=repository
```

```python
repository_data_name = repository_data["repo_url"].str.split("/", n = 4, expand = True)
repository_data_name
```

```python
repository_data_name.drop(columns=repository_data_name.columns[0], axis=1, inplace=True)
repository_data_name.drop(columns=repository_data_name.columns[0], axis=1, inplace=True)
repository_data_name.drop(columns=repository_data_name.columns[0], axis=1, inplace=True)
repository_data_name.drop(columns=repository_data_name.columns[0], axis=1, inplace=True)
repository_data_name = repository_data_name.drop_duplicates()
repository_data_name
```

```python
# loop through all pages to obtain all issues information
issues=[]
for page_num in range(10):
    for r in range(repository_data_name.shape[0]):
        repository_i = repository_data_name.iloc[r,0]
        try:
            # to find all the repos' names from each page
            url = f"https://api.github.com/repos/{repository_i}/issues?state=all&page={page_num}"
            issue=requests.get(url,headers=headers).json()
            issues.append(issue)
        except:
            issues.append(None)
```

```python
repo_url=[]
for page in issues:
    for issue in page:
        try:
            repo_url.append(issue['repository_url'])
        except:
            pass
len(repo_url)
```

```python
issue_number=[]
for page in issues:
    for issue in page:
        try:
            issue_number.append(issue['number'])
        except:
            pass
len(issue_number)
```

```python
issue_title=[]
for page in issues:
    for issue in page:
        try:
            issue_title.append(issue['title'])
        except:
            pass
len(issue_title)
```

```python
issue_body=[]
for page in issues:
    for issue in page:
        try:
            issue_body.append(issue['body'])
        except:
            pass
len(issue_body)
```

```python
issue_state=[]
for page in issues:
    for issue in page:
        try:
            issue_state.append(issue['state'])
        except:
            pass
len(issue_state)
```

```python
issue_create_time=[]
for page in issues:
    for issue in page:
        try:
            issue_create_time.append(issue['created_at'])
        except:
            pass
len(issue_create_time)
```

```python
issue_update_time=[]
for page in issues:
    for issue in page:
        try:
            issue_update_time.append(issue['updated_at'])
        except:
            pass
len(issue_update_time)
```

```python
issue_close_time=[]
for page in issues:
    for issue in page:
        try:
            issue_close_time.append(issue['closed_at'])
        except:
            pass
len(issue_close_time)
```

```python
issue_pull_request=[]
for page in issues:
    for issue in page:
        try:
            issue_pull_request.append(issue['pull_request'])
        except:
            issue_pull_request.append('not_pull_request')
len(issue_pull_request)
```

```python
issue_label=[]
for page in issues:
    for issue in page:
        try:
            issue_label.append(issue['labels'])
        except:
            pass
len(issue_label)
```

```python
issue_data=pd.DataFrame()
issue_data['repo_url']=repo_url
issue_data['issue_number']=issue_number
issue_data['issue_title']=issue_title
issue_data['issue_body']=issue_body
issue_data['issue_state']=issue_state
issue_data['issue_created_at']=issue_create_time
issue_data['issue_created_at_timestamp']=issue_create_time
issue_data['issue_updated_at']=issue_update_time
issue_data['issue_updated_at_timestamp']=issue_update_time
issue_data['issue_closed_at']=issue_close_time
issue_data['issue_closed_at_timestamp']=issue_close_time
issue_data['resolution_time']=np.nan
issue_data['issue_pull_request']=issue_pull_request
issue_data['issue_label']=issue_label
issue_data['issue_label_name']=issue_label
issue_data=issue_data.reset_index().drop(columns='index')
```

```python
j = 0
for j in range(issue_data.shape[0]):
    issue_data.iloc[j,6] = datetime.strptime(issue_data.iloc[j,5], '%Y-%m-%dT%H:%M:%SZ')
    issue_data.iloc[j,8] = datetime.strptime(issue_data.iloc[j,7], '%Y-%m-%dT%H:%M:%SZ')
    if issue_data.iloc[j,4] == 'closed':
        issue_data.iloc[j,10] = datetime.strptime(issue_data.iloc[j,9], '%Y-%m-%dT%H:%M:%SZ')
```

```python
j = 0
for j in range(issue_data.shape[0]):
    if len(issue_data.iloc[j,13]) == 0:
        issue_data.iloc[j,14] = 'no_label'
    else:
        multi_labels = []
        k = 0
        label1 = issue_data.iloc[j,13]
        for k in range(len(issue_data.iloc[j,14])):
            x = label1[k]['name']
            multi_labels += [x]
            # print(multi_labels)
            # print(k)
        issue_data.iloc[j,14] = str(multi_labels)
```
